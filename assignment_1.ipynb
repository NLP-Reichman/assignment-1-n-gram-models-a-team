# Part 1: N-gram Language Models

## Part 1.1: Preprocessing
Implement the function *preprocess* that iterates over all the data files and creates a single vocabulary, containing all the tokens in the data. Our token definition is a single UTF-8 encoded character. So, the vocabulary list is a simple Python list of all the characters that you see at least once in the data. The vocabulary should include the `<start>` and  `<end>` tokens.

Note - do NOT lowercase the sentences.

```python
def preprocess() -> list[str]:
  '''
  Return a list of characters, representing the shared vocabulary of all languages
  '''
  vocab = []
  ########################################
  chars = set()
  langs = ['en', 'es', 'fr', 'in', 'it', 'nl', 'pt', 'tl']
  for lang in langs:
    df = pd.read_csv(f'data/{lang}.csv')
    for tweet in df['tweet_text'].dropna():
      for ch in str(tweet):
        chars.add(ch)
  vocab = list(chars) + ['<start>', '<end>']
  ########################################
  return vocab
```

## Part 1.2: Building the Language Model
Implement the function *build_lm* that generates a language model from a textual corpus. The function should return a dictionary (representing a model) where the keys are all the relevant *n*-1 sequences, and the values are dictionaries with the *n*_th tokens and their corresponding probabilities to occur. To ensure consistent probabilities calculation, please add n-1 `<start>` tokens to the beginning of a tweet and one `<end>` token at the end. For example, for a trigram model (tokens are characters), it should look something like:

{ "ab":{"c":0.5, "b":0.25, "d":0.25}, "ca":{"a":0.2, "b":0.7, "d":0.1} }

which means for example that after the sequence "ab", there is a 0.5 chance that "c" will appear, 0.25 for "b" to appear and 0.25 for "d" to appear.

Note - You should think how to add the add-one smoothing information to the dictionary and implement it.

Please add the `<unk>` token with $p(<unk>)=1/|V|$ to the LM if building a smoothed LM.

```python
def build_lm(lang: str, n: int, smoothed: bool = False) -> dict[str, dict[str, float]]:
  '''
  Return a language model for the given lang and n_gram (n).

  The model is a dict of dicts:
    { context_string: { next_token: probability, ... }, ... }

  where context_string is the joined (n-1)-token prefix and next_token is
  the character (or special token) that follows it.

  Special tokens:
    <start>  — prepended (n-1) times to each tweet
    <end>    — appended once to each tweet
    <unk>    — top-level key added when smoothed=True;
               maps to {'<unk>': 1/|V|} as a fallback for unseen contexts
               during evaluation

  Smoothing (add-one / Laplace), applied when smoothed=True:
    P(c | ctx) = (count(ctx, c) + 1) / (count(ctx) + |V|)
    Unseen (ctx, c) pairs also receive 1 / (count(ctx) + |V|).

  :param lang: language code matching a file in data/<lang>.csv
  :param n: n-gram order (e.g. 2 = bigram, 3 = trigram)
  :param smoothed: whether to apply add-one smoothing
  :return: the language model dict
  '''
  LM = {}
  ########################################
  vocab = preprocess()
  V = len(vocab)

  df = pd.read_csv(f'data/{lang}.csv')

  # --- count n-gram occurrences ---
  counts = {}  # context_str -> {next_token: raw_count}
  for tweet in df['tweet_text'].dropna():
    tokens = ['<start>'] * (n - 1) + list(str(tweet)) + ['<end>']
    for i in range(len(tokens) - n + 1):
      context = ''.join(tokens[i:i + n - 1])
      next_tok = tokens[i + n - 1]
      if context not in counts:
        counts[context] = {}
      counts[context][next_tok] = counts[context].get(next_tok, 0) + 1

  # --- convert counts to probabilities ---
  for context, next_counts in counts.items():
    total = sum(next_counts.values())
    if smoothed:
      denom = total + V
      dist = {tok: (cnt + 1) / denom for tok, cnt in next_counts.items()}
      for tok in vocab:
        if tok not in dist:
          dist[tok] = 1 / denom
    else:
      dist = {tok: cnt / total for tok, cnt in next_counts.items()}
    LM[context] = dist

  # <unk> as a top-level fallback context for completely unseen contexts at eval
  if smoothed:
    LM['<unk>'] = {'<unk>': 1 / V}
  ########################################
  return LM
```

## Part 1.3: Evaluation
Implement the function `eval` that returns the perplexity of a model (dictionary) running over the data file of the given target language.

The `<unk>` should be used for unknown contexts when calculating the perplexities.

```python
def eval(model: dict, target_lang: str, n: int) -> float:
  '''
  Return the perplexity of the model evaluated on the target language's data.

  Perplexity formula:
    PP = exp( -1/N * sum( log P(c_i | context_i) ) )

  where N is the total number of tokens predicted across all tweets.

  Context lookup:
    - If context exists in model: use its distribution
    - If context is unseen: fall back to model['<unk>'] -> probability 1/|V|
    - If next token is not in the context distribution: use 1/|V|

  :param model: smoothed language model (dict of dicts) from build_lm
  :param target_lang: language code of the text to evaluate on
  :param n: n-gram order the model was built with
  :return: perplexity value (lower = better fit)
  '''
  pp = 0
  ########################################
  vocab = preprocess()
  V = len(vocab)

  df = pd.read_csv(f'data/{target_lang}.csv')

  log_prob_sum = 0.0
  N = 0

  for tweet in df['tweet_text'].dropna():
    tokens = ['<start>'] * (n - 1) + list(str(tweet)) + ['<end>']
    for i in range(len(tokens) - n + 1):
      context = ''.join(tokens[i:i + n - 1])
      next_tok = tokens[i + n - 1]

      dist = model.get(context, model.get('<unk>', {'<unk>': 1 / V}))
      prob = dist.get(next_tok, 1 / V)

      log_prob_sum += np.log(prob)
      N += 1

  pp = np.exp(-log_prob_sum / N)
  ########################################
  return pp
```

## Part 1.4: Generation
Implement the *generate* function which takes a language code, *n*, the prompt (the starting text), the number of tokens to generate, and *r*, which is the random seed for any randomized action you plan to take in your implementation. The function should start generating tokens, one by one, using the language model of the given source language and *n*. The prompt should be used as a starting point for aligning on the probabilities to be used for generating the next token.

Note - The generation of the next token should be from the LM's distribution with NO smoothing.

```python
def generate(lang: str, n: int, prompt: str, number_of_tokens: int, r: int) -> str:
  '''
  Generate text in the given language using the given parameters.

  Uses an UNSMOOTHED language model. Starting from the prompt, repeatedly:
    1. Take the last (n-1) characters of current text as context
    2. Sample the next token from the context distribution
    3. Stop early if <end> is sampled, otherwise append and continue

  If a context is not found in the LM, a random vocab token is sampled
  uniformly as fallback.

  :param lang: language code matching a file in data/<lang>.csv
  :param n: n-gram order
  :param prompt: starting text (included in the returned string)
  :param number_of_tokens: how many tokens to generate beyond the prompt
  :param r: random seed for reproducibility
  :return: prompt + generated text
  '''
  text = ""
  ########################################
  np.random.seed(r)
  lm = build_lm(lang, n, smoothed=False)
  vocab = preprocess()

  text = prompt
  for _ in range(number_of_tokens):
    context = text[-(n - 1):] if n > 1 else ''

    if context in lm:
      dist = lm[context]
      tokens = list(dist.keys())
      probs  = list(dist.values())
      next_tok = np.random.choice(tokens, p=probs)
    else:
      next_tok = np.random.choice(vocab)

    if next_tok == '<end>':
      break
    if next_tok == '<start>':
      continue

    text += next_tok
  ########################################
  return text
```

```python
########################################
# Tests

def test_link():
    return {
        'link': NOTEBOOK_LINK
    }

def test_preprocess():
    return {
        'vocab_length': len(preprocess()),
    }

def test_build_lm():
    return {
        'english_2_gram_length': len(build_lm('en', 2, True)),
        'english_3_gram_length': len(build_lm('en', 3, True)),
        'french_3_gram_length': len(build_lm('fr', 3, True)),
        'spanish_3_gram_length': len(build_lm('es', 3, True)),
    }

def test_eval():
    lm = build_lm('en', 3, True)
    return {
        'en_on_en': round(eval(lm, 'en', 3), 2),
        'en_on_fr': round(eval(lm, 'fr', 3), 2),
        'en_on_tl': round(eval(lm, 'tl', 3), 2),
        'en_on_nl': round(eval(lm, 'nl', 3), 2),
    }

def test_generate():
    return {
        'english_2_gram': generate('en', 2, "I am", 20, 5),
        'english_3_gram': generate('en', 3, "I am", 20, 5),
        'english_4_gram': generate('en', 4, "I Love", 20, 5),
        'spanish_2_gram': generate('es', 2, "Soy", 20, 5),
        'spanish_3_gram': generate('es', 3, "Soy", 20, 5),
        'french_2_gram': generate('fr', 2, "Je suis", 20, 5),
        'french_3_gram': generate('fr', 3, "Je suis", 20, 5),
    }

TESTS = [test_link, test_preprocess, test_build_lm, test_eval, test_generate]

# Run tests and save results
res = {}
for test in TESTS:
    try:
        cur_res = test()
        res.update({test.__name__: cur_res})
    except Exception as e:
        import traceback
        res.update({test.__name__: repr(e) + "\n" + traceback.format_exc()})

with open('results.json', 'w') as f:
    json.dump(res, f, indent=2)

########################################
```
